````python
# raw
def plain_text(plain_text):
    hex_form = ''.join(f"{ord(plain_text[i]):02x}" for i in range(len(plain_text)))
    byte_form = bytes.fromhex(hex_form)
    return byte_form

# key
def key_func(key):
    hex_form = ''.join(f"{ord(key[i]):02x}" for i in range(len(key)))
    byte_form = bytes.fromhex(hex_form)
    return byte_form

# xor chiper
def xor(b1, b2):
    # equalize the key
    point = 0
    edited_key = bytearray(len(b1))
    for i in range(len(edited_key)):
        if point != 2:
            edited_key[i] = b2[point]
            point += 1
        else:
            edited_key[i] = b2[point]
            point = 0
    
    # xor process
    place_holder = bytearray(min(len(edited_key), len(b1)))
    for i in range(len(place_holder)):
        place_holder[i] = edited_key[i] ^ b1[i]
    return place_holder

def repeating_xor(_byte1, _byte2):
    repeated = 0
    place_holder = []
    while repeated < 2:
        if repeated == 0:
            xor_result = xor(_byte1, _byte2)
            place_holder.append(xor_result)
        if repeated == 1:
            xor_result = xor(place_holder[0], _byte2)
            place_holder.append(xor_result)
        repeated += 1
    # print(place_holder[0])
    # print(place_holder[1])
    return place_holder

_plain_text = "Burning 'em, if you ain't quick and nimble\nI go crazy when I hear a cymbal"
_key = "ICE"

b_plain = plain_text(_plain_text)
b_key = key_func(_key)

result = repeating_xor(b_plain, b_key)
plain_result = ""
for j in result[0]:
    plain_result += f"{j:02x}"

for i in range(len(plain_result)):
    
    if i == 75:
        print()
        print(plain_result[i],end="")
    else:
        print(plain_result[i],end="")
````

````python
def repeating_key_xor(text: bytes, key: bytes) -> bytes:
    """Given a plain text `text` as bytes and an encryption key `key`
    as bytes, the function encrypts the text by performing
    XOR of all the bytes and the `key` (in repeated manner) and returns
    the resultant XORed byte stream.
    """

    # we update the encryption key by repeating it such that it
    # matches the length of the text to be processed.
    repetitions = 1 + (len(text) // len(key))
    key = key * repetitions
    key = key[:len(text)]

    # XOR text and key generated above and return the raw bytes
    return bytes([b ^ k for b, k in zip(text, key)])
````
