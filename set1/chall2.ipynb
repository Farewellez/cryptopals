## Fixed XOR
Write a function that takes two equal-length buffers and produces their XOR combination.

If your function works properly, then when you feed it the string:

```
1c0111001f010100061a024b53535009181c
```
... after hex decoding, and when XOR'd against:
```
686974207468652062756c6c277320657965
```
... should produce:
```
746865206b696420646f6e277420706c6179
```

````python
first_string = "1c0111001f010100061a024b53535009181c"
key = "686974207468652062756c6c277320657965"

bi1 = [format((int(first_string[i:i+2],16)), "08b") for i in range(0,len(first_string),2)]
biner_key = [format((int(key[i:i+2],16)), "08b") for i in range(0,len(key),2)]

result = []
for i in range(len(bi1)):
    compared = ""
    for j in range(8):
        if bi1[i][j] != biner_key[i][j]:
            compared += "1"
        else:
            compared += "0"
    result.append(compared)
    compared = ""

last = ''.join(str(hex(int(result[i],2)))[2:] for i in range(len(result)))
print(last)
````

````python
def xor(b1, b2):
    xord = bytearray(min(len(b1), len(b2)))
    for i  in range(min(len(b1), len(b2))):
        xord[i] = b1[i] ^ b2[i]
    
    return xord

r_b1 = bytes.fromhex(first_string)
r_b2 = bytes.fromhex(key)
print(r_b1,end=f"\n{r_b2}\n") # this is the chiper and the key

print(xor(r_b1, r_b2)) # this the decrypt
````
