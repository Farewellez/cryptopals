## Convert hex to base64

The string:
```
49276d206b696c6c696e6720796f757220627261696e206c696b65206120706f69736f6e6f7573206d757368726f6f6d
```

Should produce:
```
SSdtIGtpbGxpbmcgeW91ciBicmFpbiBsaWtlIGEgcG9pc29ub3VzIG11c2hyb29t
```

### Cryptopals Rule
Always operate on raw bytes, never on encoded strings. Only use hex and base64 for pretty-printing.

````python
from string import ascii_lowercase, ascii_uppercase, digits
````

````python
base64_list = ascii_uppercase + ascii_lowercase + digits + "+/"

raw_string = "49276d206b696c6c696e6720796f757220627261696e206c696b65206120706f69736f6e6f7573206d757368726f6f6d"
target_string = "SSdtIGtpbGxpbmcgeW91ciBicmFpbiBsaWtlIGEgcG9pc29ub3VzIG11c2hyb29t"


biner_form = [str(int(raw_string[i:i+2],16)) for i in range(0, len(raw_string), 2)]
_8digit = ''.join(format(int(biner_form[i]), "08b") for i in range(len(biner_form)))
_6digit = ''.join((base64_list[int(_8digit[i:],2)] if len(_8digit[i:]) % 6 == 0 else base64_list[int((_8digit[i:]) + '0'*(6 - len(_8digit[i:])),2)]) if len(_8digit) - i <= 6 else base64_list[int(_8digit[i:i+6], 2)] for i in range(0, len(_8digit), 6))
result = ''.join(_6digit if len(_6digit)%4 == 0 else _6digit + '='*(len(_6digit)%4))
print(f"{result}: SUCCES!!!" if result == target_string else f"{result}: FAILED!!!")
````

````python
import base64
````

````python
def hex_to_bytes(s):
    return bytes.fromhex(s)

def bytes_to_base64(b):
    return base64.b64encode(b)

hex_bytes = hex_to_bytes(raw_string)
b64_bytes = bytes_to_base64(hex_bytes)
print(b64_bytes.decode(encoding="utf-8"))
````
